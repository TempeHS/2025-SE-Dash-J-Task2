#### Line Plot with Rolling Average
Purpose: To smooth out fluctuations and observe long-term trends in NSW_total.
Reasoning: A rolling average helps identify trends by reducing noise in the data

```python
# Define the columns to plot
columns_to_plot = ['NSW_total', 'VIC_total', 'QLD_total', 'SA_total', 'WA_total', 'TAS_total', 'ACT_total']

# Calculate rolling averages for each state
rolling_avg = test_data[columns_to_plot].rolling(window=5).mean()

# Plot each state's total and its rolling average
plt.figure(figsize=(12, 8))
for column in columns_to_plot:
    plt.plot(test_data['YEAR'], test_data[column], label=f'{column} Total')
    plt.plot(test_data['YEAR'], rolling_avg[column], label=f'{column} Rolling Avg', linestyle='--')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Total Value')
plt.title('State Totals with Rolling Averages')
plt.legend()
plt.show()
```
